```python
# Adivinhe a palavra!
import random
palavras=['Computador', 'Bolacha', 'Monitor', 'Teclado', 'Mouse', 'Python', 'Javascript']
palavras_de_animo=['Bahhh, você não acertou dessa vez.', 'Na próxima vai', 'Essa foi quase hein.', 'Pensa bem viu.', 'Você consegue, eu confio!']

def palavra():
  i=random.randrange(len(palavras))
  palavra=palavras[i]
  palavra=str.lower(palavra)
  return palavra

def palavra_de_animo():
  palavra_de_animo=random.choice(palavras_de_animo)
  return palavra_de_animo

def embaralhar_palavra(palavra):
  palavra_lista=list(palavra)
  random.shuffle(palavra_lista)
  palavra_embaralhada=''.join(palavra_lista)
  return palavra_embaralhada

def jogo():
  tentativas=5
  tentativa_i=0
  palavra_certa=palavra()
  print(f"""
  Bem vindo ao O JOGO!!

  Você tem 5 tentativas para acertar a palavra embaralhada. Boa sorte!!
  A palavra embaralhada é *{embaralhar_palavra(palavra_certa)}*
  """)
  while tentativas>0:
    tentativa=str.lower(input(f"Digite sua {tentativa_i+1}º tentativa: "))
    tentativa_i+=1
    tentativas-=1
    if tentativa==palavra_certa:
      print(f"\nParabéns!!! Você acertou em {tentativa_i} tentativas!!!")
      break
    else:
      print(palavra_de_animo())
      print(f"Você ainda tem {tentativas} tentativas\n")
  if tentativas==0:
    print(f"\nVocê perdeu :(, a palavra certa era {palavra_certa}")
jogo()
```

```python
# Adivinhe a palavra!
import random
temas = {
    "Países": {
        "Curtas": ["Japão", "Chile", "Malta", "Haiti", "Suíça"],
        "Médias": ["Brasil", "França", "Itália", "Canadá", "Índia"],
        "Longas": ["Portugal", "Venezuela", "Argentina", "Filipinas", "República Tcheca"]
    },
    "Cores": {
        "Curtas": ["Azul", "Rosa", "Cinza", "Marrom", "Ouro"],
        "Médias": ["Verde", "Amarelo", "Laranja", "Rosa-choque", "Verde-água"],
        "Longas": ["Violeta", "Turquesa", "Coral", "Amarelo-queimado", "Azul-celeste"]
    },
    "Times": {
        "Curtas": ["Flamengo", "Chelsea", "Lyon", "PSG"],
        "Médias": ["Barcelona", "Juventus", "Arsenal", "Inter", "Roma"],
        "Longas": ["Manchester City", "Borussia Dortmund", "Atlético Mineiro", "Real Madrid", "Manchester United"]
    },
    "Objetos": {
        "Curtas": ["Mesa", "Caneta", "Livro", "Chave", "Porta"],
        "Médias": ["Telefone", "Computador", "Cadeira", "Caderno", "Garrafa"],
        "Longas": ["Refrigerador", "Micro-ondas", "Aspirador de pó", "Relógio inteligente", "Máquina de lavar"]
    },
    "Tecnologia": {
        "Curtas": ["App", "USB", "Wi-Fi", "DVD", "VPN"],
        "Médias": ["Internet", "Smartphone", "Software", "Impressora", "Monitor"],
        "Longas": ["Computador portátil", "Inteligência artificial", "Realidade aumentada", "Sistema operacional", "Banco de dados"]
    }
}
palavras_de_animo=['Bahhh, você não acertou dessa vez.', 'Na próxima vai', 'Essa foi quase hein.', 'Pensa bem viu.', 'Você consegue, eu confio!']
def tema():
  tema=int(input("""Escolha o tema:
  1- Países
  2- Cores
  3- Times
  4- Objetos
  5- Tecnologia\n"""))
  return tema

def dificuldade():
  dificuldade=int(input("""Escolha a dificuldade:
  1- Fácil
  2- Médio
  3- Díficil\n"""))
  return dificuldade

def palavra():
  tema_escolha=tema()
  dificuldade_escolha=dificuldade()
  tema_str=''
  if tema_escolha==1:
    tema_str='Países'
  elif tema_escolha==2:
    tema_str='Cores'
  elif tema_escolha==3:
    tema_str='Times'
  elif tema_escolha==4:
    tema_str='Objetos'
  elif tema_escolha==5:
    tema_str='Tecnologia'

  if dificuldade_escolha == 1:
    dificuldade_str='Curtas'
  elif dificuldade_escolha == 2:
    dificuldade_str='Médias'
  elif dificuldade_escolha == 3:
    dificuldade_str= 'Longas'
  palavra=(random.choice(temas[tema_str][dificuldade_str]))
  palavra=str.lower(palavra)
  return palavra

def palavra_de_animo():
  palavra_de_animo=random.choice(palavras_de_animo)
  return palavra_de_animo

def embaralhar_palavra(palavra):
  palavra_lista=list(palavra)
  random.shuffle(palavra_lista)
  palavra_embaralhada=''.join(palavra_lista)
  return palavra_embaralhada

def jogo():
  tentativas=5
  tentativa_i=0
  palavra_certa=palavra()
  print(f"""
  \nBem vindo ao O JOGO!!

  Você tem 5 tentativas para acertar a palavra embaralhada. Boa sorte!!
  A palavra embaralhada é *{embaralhar_palavra(palavra_certa)}*
  """)
  while tentativas>0:
    tentativa=str.lower(input(f"Digite sua {tentativa_i+1}º tentativa: "))
    tentativa_i+=1
    tentativas-=1
    if tentativa==palavra_certa:
      print(f"\nParabéns!!! Você acertou em {tentativa_i} tentativas!!!")
      break
    else:
      print(palavra_de_animo())
      print(f"Você ainda tem {tentativas} tentativas\n")
  if tentativas==0:
    print(f"\nVocê perdeu :(, a palavra certa era {palavra_certa}")
jogo()
```
