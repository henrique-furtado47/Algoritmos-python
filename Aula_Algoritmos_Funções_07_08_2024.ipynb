**1. Faça uma função para converter uma temperatura em graus Fahrenheit para
Celsius. A temperatura em Fahrenheit é o dado de entrada e a temperatura em
Celsius é o dado de saída. Utilize a fórmula C = (F - 32) * 5/9, onde F é a
temperatura em Fahrenheit e C é a temperatura em Celsius.**

```python
def conversao(fahrenheit):
  celsius=(fahrenheit-32)*(5/9)
  return celsius

temp=float(input("Digite a temperatura em Fahrenheit: "))
temp_c=conversao(temp)
print(f"A temperatura de {temp}°F em Celsius é : {temp_c}°C")
```

```python
import math
def calcular_hipotenusa(catetoA, catetoB):
  hipotenusa = math.sqrt(catetoA**2+catetoB**2)
  return hipotenusa
cat1=float(input("Digite o valor do cateto A: "))
cat2=float(input("Digite valor do cateto B: "))
resultado=calcular_hipotenusa(cat1,cat2)
print(f"A hipotenusa é aproximadamente: {resultado:.2f}")
```

**3. Escreva um programa para ler as notas das duas avaliações de um aluno no
semestre. Faça uma função que receba as duas notas por parâmetro e calcule e
escreva a média semestral e a mensagem “PARABÉNS! Você foi aprovado!”
somente se o aluno foi aprovado (considere 6.0 a média mínima para aprovação).**

```python
def calcular_media(nota1, nota2):
  media=(nota1+nota2)/2
  if media>=6:
    msg="PARABÉNS! Você foi aprovado"
  else:
    msg="REPROVADO"
  return msg

notaA=float(input("Digite a 1º nota do aluno: "))
notaB=float(input("Digite a 2º nota do aluno: "))
resultado=calcular_media(notaA,notaB)
print(resultado)
```

**6. Escreva uma função que recebe 2 números inteiros n1 e n2 como entrada e
retorna a soma de todos os números inteiros contidos no intervalo [n1,n2]. Use
esta função em um programa que lê n1 e n2 do usuário e imprime a soma.**

```python
def calcular_intervalo(n1,n2):
  intervalo=list(range(n1,n2))
  soma=sum(intervalo)-n1
  return soma

numero1=int(input("Digite o valor do número 1: "))
numero2=int(input("Digite o valor do número 2: "))
resultado=calcular_intervalo(numero1,numero2)
print(f"\nO resultado da soma do intervalo entre os números {numero1} e {numero2} é: {resultado}")
```

**7. Escreva uma função que receba um número inteiro e imprima o mês
correspondente ao número. Por exemplo, 2 corresponde a “fevereiro”. O
procedimento deve mostrar uma mensagem de erro caso o número recebido não
faça sentido. Gere também um programa que leia um valor e chame a função
criada.**

```python
def saber_mes(numero):
  meses=["Janeiro","Fevereiro","Março","Abril","Maio","Junho","Julho","Agosto","Setembro","Outubro","Novembro","Dezembro"]
  if numero>12 or numero<1:
    print("Valor Inválido")
  else:
    print(f"\nMês {numero} = {meses[numero-1]}")
valor=int(input("Digite o número do mês: "))
saber_mes(valor)
```

**9. Escreva uma função que receba dois números inteiros x e y. Essa função deve
verificar se x é divisível por y. No caso positivo, a função deve retornar 1, caso
contrário zero. Escreva também um programa para testar tal função.**

```python
def descobrir_divisivel(x, y):
  if x%y==0:
    return 1
  else:
    return 0

valorx=(int(input("Digite o valor de x: ")))
valory=(int(input("Digite o valor y: ")))
descobrir_divisivel(valorx,valory)
```

**10. Criar uma função que calcule e retorne o MAIOR entre dois valores recebidos como
parâmetros. Um programa para testar tal função deve ser criado.**

```python
def calcular_maior(x,y):
  if x>y:
    return x
  elif x<y:
    return y
valorx=(int(input("Digite o valor do primeiro número: ")))
valory=int(input("Digite o valor do segundo número: "))

if valorx==valory:
  print("Os números são iguais")
else:
  print(f"O maior número é {calcular_maior(valorx, valory)}")
```
