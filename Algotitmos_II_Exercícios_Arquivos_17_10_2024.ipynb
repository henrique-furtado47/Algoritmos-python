```python
arquivo_original=open('texto.txt', 'r')
arquivo_copia=open('copia.txt', 'r')
arquivo_combinado=open('combinado.txt', 'w')

conteudo1=arquivo_original.read()
conteudo2=arquivo_copia.read()

arquivo_combinado.write(conteudo1 + conteudo2))
arquivo_combinado.close()
arquivo_original.close()
arquivo_copia.close()
```
