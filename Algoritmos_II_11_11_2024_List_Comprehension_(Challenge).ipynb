```python
matriz=[[1, 2, 3],
        [4, 5, 6],
        [7, 8, 9]]

matriz_transposta = [[linha[i] for linha in matriz] for i in range(len(matriz[0]))]
for linha in matriz_transposta:
  print(linha)
```

```python
def is_prime(numero):
  if numero<2:
    return False
  for i in range(2, int(numero**0.5) + 1):
    if numero%i==0:
      return False
  return True

lista=[(i, i**2, i**3) for i in range(1, 21) if is_prime(i)]
print(lista)
```

```python
string="amongus"
sub_strings= [string[i:j] for i in range(len(string)) for j in range(i + 1, len(string) + 1)]
print(sub_strings)
```

```python

palavras = [
    "Casa", "Amizade", "Esperança", "Coragem", "Alegria", "Viagem", "Sonho",
    "Natureza", "Sabedoria", "Luz", "Paz", "Liberdade", "Amor", "Inspiração",
    "Criatividade", "Aprendizado", "Harmonia", "Confiança", "Felicidade", "Descoberta"
]

anagramas=[''.join(sorted(palavra.lower())) for palavra in palavras]
print(f"Anagramas: {anagramas}")
```
