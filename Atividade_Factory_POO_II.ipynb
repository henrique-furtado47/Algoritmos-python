```python
from abc import ABC, abstractmethod

# Interface do Produto (Product)
class MotorVehicle(ABC):
    @abstractmethod
    def build(self) -> None:
        pass

# Produtos Concretos (Concrete Products)
class Motorcycle(MotorVehicle):
    def build(self) -> None:
        print("Construindo uma Motocicleta")

class Car(MotorVehicle):
    def build(self) -> None:
        print("Construindo um Carro")

# Criador (Creator)
class MotorVehicleFactory(ABC):
    @abstractmethod
    def create_motor_vehicle(self) -> MotorVehicle:
        pass

    def create(self) -> MotorVehicle:
        # Lgica adicional pode ser inserida aqui antes da criao.
        vehicle = self.create_motor_vehicle()
        vehicle.build()
        return vehicle

# Criadores Concretos (Concrete Creators)
class MotorcycleFactory(MotorVehicleFactory):
    def create_motor_vehicle(self) -> MotorVehicle:
        return Motorcycle()

class CarFactory(MotorVehicleFactory):
    def create_motor_vehicle(self) -> MotorVehicle:
        return Car()

# Cdigo cliente
def client_code(factory: MotorVehicleFactory):
    print(f"Cliente: Eu no conheo a classe do criador, mas funciona mesmo assim.")
    product = factory.create()

print("App: Lanado com a MotorcycleFactory.")
client_code(MotorcycleFactory())

print("\nApp: Lanado com a CarFactory.")
client_code(CarFactory())
```

```python
from abc import ABC, abstractmethod

# Interfaces dos Produtos
class Armchair(ABC):
    @abstractmethod
    def sit_on(self):
        pass

class CoffeeTable(ABC):
    @abstractmethod
    def put_on(self):
        pass

class Sofa(ABC):
    @abstractmethod
    def lie_on(self):
        pass

# Produtos Concretos (Ex: Estilo Moderno)
class ModernArmchair(Armchair):
    def sit_on(self):
        return "Sentando em uma poltrona moderna."

class ModernCoffeeTable(CoffeeTable):
    def put_on(self):
        return "Colocando algo na mesa de centro moderna."

class ModernSofa(Sofa):
    def lie_on(self):
        return "Deitando em um sof moderno."

# Produtos Concretos (Ex: Estilo Retr)
class RetroArmchair(Armchair):
    def sit_on(self):
        return "Sentando em uma poltrona retr."

class RetroCoffeeTable(CoffeeTable):
    def put_on(self):
        return "Colocando algo na mesa de centro retr."

class RetroSofa(Sofa):
    def lie_on(self):
        return "Deitando em um sof retr."

# Criador Abstrato (Abstract Creator / Abstract Factory)
class FurnitureFactory(ABC):
    @abstractmethod
    def make_armchair(self) -> Armchair:
        pass

    @abstractmethod
    def make_coffee_table(self) -> CoffeeTable:
        pass

    @abstractmethod
    def make_sofa(self) -> Sofa:
        pass

# Criadores Concretos (Concrete Creators / Concrete Factories)
class ModernFurnitureFactory(FurnitureFactory):
    def make_armchair(self) -> Armchair:
        return ModernArmchair()

    def make_coffee_table(self) -> CoffeeTable:
        return ModernCoffeeTable()

    def make_sofa(self) -> Sofa:
        return ModernSofa()

class RetroFurnitureFactory(FurnitureFactory):
    def make_armchair(self) -> Armchair:
        return RetroArmchair()

    def make_coffee_table(self) -> CoffeeTable:
        return RetroCoffeeTable()

    def make_sofa(self) -> Sofa:
        return RetroSofa()

# Cdigo Cliente
def client_code_furniture(factory: FurnitureFactory):
    armchair = factory.make_armchair()
    coffee_table = factory.make_coffee_table()
    sofa = factory.make_sofa()

    print(armchair.sit_on())
    print(coffee_table.put_on())
    print(sofa.lie_on())

print("Cliente: Testando com a fbrica de mveis modernos.")
client_code_furniture(ModernFurnitureFactory())

print("\nCliente: Testando com a fbrica de mveis retr.")
client_code_furniture(RetroFurnitureFactory())
```
