```python
import string

def verificar_digitos(text):
  for digit in text:
    if digit not in string.digits:
      return False
  return True

print(verificar_digitos(str(input("Digite uma string para verificar se contém apenas digitos: "))))
```

```python
import numpy as np

def criar_matriz(linhas, colunas, conteudo=0):
  matriz=[]
  for i in range(linhas):
    linha=[]
    for j in range(colunas):
      linha.append(conteudo)
    matriz.append(linha)
  return np.array(matriz)

print(criar_matriz(3, 3))
```

```python
from bs4 import BeautifulSoup
import requests

url= 'https://balaiocientifico.com/python/selecionar-linha-especifica-no-pandas/'
response=requests.get(url)

if response.status_code == 200:
  soup= BeautifulSoup(response.content, 'html.parser')
  print("Títulos encontrados na página:")
  for title in soup.find_all('h1'):
    print(title.get_text(strip=True))
  print("\nLinks encontrados na página:")
  for link in soup.find_all('a', href=True):
    print(f"{link.get_text(strip=True)}: {link['href']}")
else:
    print(f"Falha ao acessar a página. Código de status HTTP: {response.status_code}")
```
