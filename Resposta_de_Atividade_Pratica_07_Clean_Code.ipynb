### **Exercício 1: Nomes Significativos**

> Nomes devem ser claros, precisos e consistentes. Evite usar nomes genéricos, abreviações, números mágicos ou comentários para explicar o que o código faz.

#### **Problema**

Escreva uma função que calcule o salário líquido de um funcionário a partir do seu salário bruto e dos impostos aplicados. Use nomes significativos para as variáveis e a função.

##### **Código Ruim**

```python
def f(x):
  y = 0.15 * x  # INSS
  z = 0.275 * x  # IRPF
  return x - y - z

# Example usage:
sb = 10000
sl = f(sb)
print(sl)
```

##### **Código Melhorado**

```python
def aplicar_impostos(salario_sem_impostos):
  inss = 0.15 * salario_sem_impostos
  irpf = 0.275 * salario_sem_impostos
  return salario_sem_impostos - inss - irpf

salario_bruto = 10000
salario_liquido = aplicar_impostos(salario_bruto)
print(salario_liquido)
```

### **Exercício 3: Funções (Métodos)**

> Escreva funções que sejam pequenas, simples e façam apenas uma coisa. Use nomes descritivos para as funções e os parâmetros. Opcionalmente, use `docstrings` para documentar o propósito, os parâmetros e o retorno das funções.

#### **Problema**

Escreva uma função que efetue as quatro operações aritméticas (i.e. soma, subtração, divisão e multiplição), recebendo como parâmetro os dois números e a operação desejada (`+`, `-`, `*` e `/`) em forma de símbolos.

##### **Código Ruim**

```python
def calc(x, y, op):
  # faz uma operação entre dois números
  if op == "+":
    return x + y
  elif op == "-":
    return x - y
  elif op == "*":
    return x * y
  elif op == "/":
    return x / y
  else:
    return "Operação inválida"
```

##### **Código Melhorado**

```python
def calcular(num1, num2, simbolo):
  """
  Realiza uma operação aritmética entre dois números.

  Args:
    num1: O primeiro número.
    num2: O segundo número.
    simbolo: O símbolo da operação desejada ('+', '-', '*', '/').

  Returns:
    O resultado da operação ou uma mensagem de erro se a operação for inválida ou divisão por zero.
  """
  match simbolo:
    case '+':
      return num1 + num2
    case '-':
      return num1 - num2
    case '*':
      return num1 * num2
    case '/':
      if num2 != 0:
        return num1 / num2
      else:
        return "Erro: Divisão por zero!"
    case _:
      return "Operação inválida"

# Exemplos de uso:
print(f"10 + 5 = {calcular(10, 5, '+')}")
print(f"10 - 5 = {calcular(10, 5, '-')}")
print(f"10 * 5 = {calcular(10, 5, '*')}")
print(f"10 / 5 = {calcular(10, 5, '/')}")
print(f"10 / 0 = {calcular(10, 0, '/')}")
print(f"10 % 5 = {calcular(10, 5, '%')}")
```

#### **Problema**

Escreva um programa que armazene as oito notas e os nomes dos estudantes da disciplina de POO II. Calcule a média da turma e imprima os nomes e as notas daqueles ficaram acima da média.

##### **Código Ruim**

```python
# armazena os nomes e as notas dos alunos em uma sequência
alunos = "Ana", "Bruno", "Carla", "Daniel", "Eduardo", "Fernanda", "Gabriel", "Helena"
notas = 8.5, 7.0, 9.0, 6.5, 10.0, 8.0, 7.5, 9.5

# calcula a média das notas dos alunos
soma = 0
for i in range(len(notas)):
  soma += notas[i]
media = soma / len(notas)

# imprime os nomes e as notas dos alunos que ficaram acima da média
for i in range(len(alunos)):
  if notas[i] > media:
      print(alunos[i], notas[i])
```

##### **Código Melhorado**

```python
alunos_notas = [
    {"nome": "Ana", "nota": 8.5},
    {"nome": "Bruno", "nota": 7.0},
    {"nome": "Carla", "nota": 9.0},
    {"nome": "Daniel", "nota": 6.5},
    {"nome": "Eduardo", "nota": 10.0},
    {"nome": "Fernanda", "nota": 8.0},
    {"nome": "Gabriel", "nota": 7.5},
    {"nome": "Helena", "nota": 9.5}
]

soma_notas = sum(aluno["nota"] for aluno in alunos_notas)
media_turma = soma_notas / len(alunos_notas)

print(f"Média da turma: {media_turma:.2f}")
print("Alunos acima da média:")

for aluno in alunos_notas:
  if aluno["nota"] > media_turma:
    print(f"{aluno["nome"]} (Nota: {aluno["nota"]})")
```

##### **Código Ruim**

```python
def soma_quadrados(lista): return sum(x**2 for x in lista) # retorna a soma dos quadrados dos elementos da lista

def media(lista): return sum(lista)/len(lista) # retorna a média dos elementos da lista

def variancia(lista): m = media(lista); return soma_quadrados([x-m for x in lista])/len(lista) # retorna a variância dos elementos da lista

def desvio_padrao(lista): return variancia(lista)**0.5 # retorna o desvio padrão dos elementos da lista
```

##### **Código Melhorado**

```python
import math

def soma_quadrados(lista):
    return sum(x**2 for x in lista)


def media(lista):
    return sum(lista) / len(lista)


def variancia(lista):
    m = media(lista)
    return sum((x - m) ** 2 for x in lista) / len(lista)


def desvio_padrao(lista):
    return math.sqrt(variancia(lista))

# Exemplo de uso:
dados = [1, 2, 3, 4, 5]
print(f"Soma dos quadrados: {soma_quadrados(dados)}")
print(f"Média: {media(dados)}")
print(f"Variância: {variancia(dados)}")
print(f"Desvio Padrão: {desvio_padrao(dados)}")

dados_vazios = []
try:
    print(f"Média de dados vazios: {media(dados_vazios)}")
except ZeroDivisionError as e:
    print(f"Erro ao calcular a média: {e}")
```

##### **Código Melhorado**

```python
class Pessoa:
    def __init__(self, nome, idade, cidade, pais, profissao, email):
        self.nome = nome
        self.idade = idade
        self.cidade = cidade
        self.pais = pais
        self.profissao = profissao
        self.email = email

    def __str__(self):
        return f"Nome: {self.nome}, Idade: {self.idade}, Cidade: {self.cidade}, País: {self.pais}, Profissão: {self.profissao}, Email: {self.email}"

# Função melhorada usando a classe Pessoa
def criar_pessoa(dados_pessoa):
    # 'dados_pessoa' pode ser um dicionário ou um objeto de uma classe customizada
    # Aqui, assumimos que 'dados_pessoa' é uma instância da classe Pessoa
    if isinstance(dados_pessoa, Pessoa):
        return dados_pessoa
    else:
        # Caso receba um dicionário, pode-se criar uma instância da classe Pessoa
        return Pessoa(
            dados_pessoa.get('nome'),
            dados_pessoa.get('idade'),
            dados_pessoa.get('cidade'),
            dados_pessoa.get('pais'),
            dados_pessoa.get('profissao'),
            dados_pessoa.get('email')
        )

# Exemplo de uso com a classe Pessoa
usuario1 = Pessoa("Alice", 30, "São Paulo", "Brasil", "Engenheira", "alice@email.com")
print("Usuário 1 (classe):", usuario1)

# Exemplo de uso passando um dicionário (que a função converte para a classe)
dados_usuario2 = {
    "nome": "Bob",
    "idade": 25,
    "cidade": "Rio de Janeiro",
    "pais": "Brasil",
    "profissao": "Designer",
    "email": "bob@email.com"
}
usuario2 = criar_pessoa(dados_usuario2)
print("Usuário 2 (dicionário convertido):", usuario2)
```
