```python
numeros=[]
nomes=[]
arquivo=open('usuarios.txt', 'r')
conteudo=arquivo.read()
texto=conteudo.split()
for i in range(len(texto)):
  if (i+1)%2==0:
    numeros.append(int(texto[i]))
  else:
    nomes.append(texto[i])

def converter_mb(numeros):
  numeros_convertidos=[]
  for i in numeros:
    megabytes=i/1e+6
    numeros_convertidos.append(f"{megabytes:.2f}")
  return numeros_convertidos

def calcular_porcentagem(numeros):
  soma=sum(numeros)
  porcentagem=[]
  for numero in numeros:
    porcentagem.append(f"{(numero/soma)*100:.2f}%")
  return porcentagem

total=sum([float(elem) for elem in converter_mb(numeros)])
media=total/len(numeros)
with open('relatorio.txt', 'w') as relatorio:
  relatorio.write('ACME Inc.\tUso do espaço em disco pelos usuarios\n')
  relatorio.write('-----------------------------------------------------------------\n')
  relatorio.write('Nr.\t  Usuario\t     Espaço utilizado\t      % do uso\n')
  relatorio.write('')
  for i in range(len(nomes)):
    relatorio.write(f"{i+1}\t   {nomes[i]}\t           {converter_mb(numeros)[i]} MB\t      {calcular_porcentagem(numeros)[i]}\n")
  relatorio.write(f'\nEspaço total ocupado: {total: .2f} MB')
  relatorio.write(f'\nEspaço médio ocupado: {media: .2f} MB')
arquivo=open('relatorio.txt', 'r')
conteudo=arquivo.read()
arquivo.close()
print(conteudo)
```
