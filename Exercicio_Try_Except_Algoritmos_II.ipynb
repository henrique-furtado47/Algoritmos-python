```python
def divisao(x, y):
    try:
      return x / y
    except ZeroDivisionError:
      return "Divisão por zero não permitida"
print(divisao(10, 2))
print(divisao(10, 0))
```

```python
from datetime import datetime
def verificar_data(date):
  try:
    datetime.strptime(date, '%d/%m/%Y')
    return "Data Válida"
  except ValueError:
    return "Data inválida"

data=str(input("Digite uma data no formato (dd/mm/yyyy): "))
print(verificar_data(data))
data=str(input("Digite uma data no formato (dd/mm/yyyy): "))
print(verificar_data(data))
```

```python
def soma(lista):
  try:
    return sum(lista)
  except TypeError:
    return "Lista inválida"

lista1=[1,2,3,4,5,6,7]
print(soma(lista1))

lista2=[1,2,3,'4']
print(soma(lista2))
```

```python
def ler_arquivo(nome):
  try:
    return nome.open('r')
  except AttributeError:
    return "Arquivo Inválido"

print(ler_arquivo('jorge'))
```

```python
def fatorial(num):
  try:
    assert num>=0 and type(num)==int
    fat = 1
    i = 2
    while i <= num:
      fat = fat*i
      i = i + 1
    return fat
  except AssertionError:
    return 'Argumento Inválido'

print(fatorial(4))
print(fatorial(5))
print(fatorial(3.2))
print(fatorial(-1))
```

```python
def contar_vogais(string):
  quantidade=0
  vogais=['a','e','i','o','u']
  try:
    assert type(string)==str
    for i in string:
      if i in vogais:
        quantidade+=1
    return quantidade
  except:
    return "Argumento Inválido"

print(contar_vogais("Eu gosto de batata"))
print(contar_vogais(3))
```

```python
def ordernar(string):
  try:
    for i in string:
      if type(i)!=str:
        raise 'ElementError'
    string.sort()
    return string
  except:
    return 'Lista Inválida'

print(ordernar(['b','d','r','f']))
print(ordernar([1, 'f']))
```

```python
def selecionar_item(lista,indice):
  try:
    return lista[indice]
  except:
    return "Indice Inválido"

a=list(range(100))
print(selecionar_item(a, 200))
```
