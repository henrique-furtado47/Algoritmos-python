```python
def divisao(x, y):
    try:
      return x / y
    except ZeroDivisionError:
      return "Divisão por zero não permitida"
print(divisao(10, 2))
print(divisao(10, 0))
```

```python
from datetime import datetime
def verificar_data(date):
  try:
    datetime.strptime(date, '%d/%m/%Y')
    return "Data Válida"
  except ValueError:
    return "Data inválida"

data=str(input("Digite uma data no formato (dd/mm/yyyy): "))
print(verificar_data(data))
data=str(input("Digite uma data no formato (dd/mm/yyyy): "))
print(verificar_data(data))
```

```python
def soma(lista):
  try:
    return sum(lista)
  except TypeError:
    return "Lista inválida"

lista1=[1,2,3,4,5,6,7]
print(soma(lista1))

lista2=[1,2,3,'4']
print(soma(lista2))
```

```python
def abrir_arquivo(nome):
  try:
    return nome.open('r')
  except AttributeError:
    return "Arquivo Inválido"

print(abrir_arquivo('jorge'))
```
